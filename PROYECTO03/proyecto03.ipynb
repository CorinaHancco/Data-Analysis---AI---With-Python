## Agrupando datos

```python
# cuanto se vendio en cada tienda?
data.groupby("tienda")["precio"].sum()
```

```python
# cuanto se vendio en cada tienda?
data.groupby("tienda")["precio"].sum().to_frame() # Para ponerlo en tabla
```

```python
data.groupby("forma_pago")["precio"].sum().to_frame()
```

```python
data_agrupada = data.groupby(["tienda", "forma_pago"])["precio"].sum().to_frame()
data_agrupada
```

```python
data_agrupada.to_excel("facturacion.xlsx")
```

## Visualización de datos

```python
import plotly.express as px
```

```python
grafico = px.histogram(data,
                       x = "tienda",
                       text_auto=True,
                       title="Facturación",
                       color="forma_pago")
grafico.show()
grafico.write_html("facturacion.html")
```
